```python
import numpy as np
import matplotlib.pyplot as plt


it=0
while (it<100):
    #Get the top 100 Gotchis out of 7782 random portals
    n_samples, n_draws, n_traits = 7782, 10, 6
    rand_traits_vals = np.random.randint(100, size=(n_samples, n_draws, n_traits))
    traits_scores = np.abs(rand_traits_vals - 49.5) + 50.5 # score = x if > 50 else 100-x
    gotchis_brs = np.sum(traits_scores, axis=2)
    max_of_draws = np.max(gotchis_brs, axis=1)
    top_100_idx_h1 = np.argsort(max_of_draws)[-100:]
    top_100_values_h1 = [max_of_draws[i] for i in top_100_idx_h1]

    #Get the top 100 Gotchis out of 7717 random portals
    n_samples, n_draws, n_traits = 7717, 10, 6
    rand_traits_vals = np.random.randint(100, size=(n_samples, n_draws, n_traits))
    traits_scores = np.abs(rand_traits_vals - 49.5) + 50.5 # score = x if > 50 else 100-x
    gotchis_brs = np.sum(traits_scores, axis=2)
    max_of_draws = np.max(gotchis_brs, axis=1)
    top_100_idx_h2 = np.argsort(max_of_draws)[-100:]
    top_100_values_h2 = [max_of_draws[i] for i in top_100_idx_h2]


    brs = 600
    rank=1
    ranks=[]
    while (brs>500):
        for h1_gotchi in top_100_values_h1:
            if (h1_gotchi==brs):
                ranks.append("h1")
                rank+=1
        for h2_gotchi in top_100_values_h2:
            if (h2_gotchi==brs):
                ranks.append("h2")
                rank+=1
        brs-=1
        
    h1s=0
    h2s=0
    
    for i in range (0,len(ranks)):
        if (ranks[i] == "h1"):
            h1s += 1
        elif (ranks[i] == "h2"):
            h2s += 1
        if (i == 9):
            print(f' Iteration {it}')
            print (f' Top 10 = {h1s} H1 + {h2s} H2')
        elif (i == 49):
            print (f' Top 50 = {h1s} H1 + {h2s} H2')
        elif (i == 99):
            print (f' Top 100 = {h1s} H1 + {h2s} H2')
    it+=1
```
